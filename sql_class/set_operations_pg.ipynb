#### Intersections

- **Intersection** --something in common
- **disjoint** -- nothing in common

````python
query = """
SELECT givenname, familyname
FROM customers
INTERSECT
SELECT givenname, familyname
FROM employees;
"""
````

````python
emp_who_are_cust = pd.read_sql(query, engine)
````
